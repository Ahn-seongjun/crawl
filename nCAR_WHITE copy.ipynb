```python
# WHITE
car_brand = []
car_name = []
car_info = []
car_year = []
car_km = []
car_location = []
car_price = []
G_D = []

for i in range (1,483):
    driver = webdriver.Chrome(executable_path="C:\chromedriver.exe")
    driver.get(f'http://www.encar.com/dc/dc_carsearchlist.do?carType=kor&searchType=model&TG.R=A#!%7B%22action%22%3A%22(And.Hidden.N._.CarType.Y._.(Or.Color.%ED%9D%B0%EC%83%89._.Color.%ED%9D%B0%EC%83%89%ED%88%AC%ED%86%A4.)_.SellType.%EC%9D%BC%EB%B0%98.)%22%2C%22toggle%22%3A%7B%7D%2C%22layer%22%3A%22%22%2C%22sort%22%3A%22ModifiedDate%22%2C%22page%22%3A[{i}]%2C%22limit%22%3A%2250%22%7D')
    time.sleep(4)
    
    for j in range(1,52):
        car_Brand = driver.find_element_by_xpath(f'//*[@id="sr_normal"]/tr[{j}]/td[2]/a/span[1]/strong')
        car_brand.append(car_Brand.text)
        time.sleep(0.2)
        
        car_Name = driver.find_element_by_xpath(f'//*[@id="sr_normal"]/tr[{j}]/td[2]/a/span[1]/em')
        car_name.append(car_Name.text)
        time.sleep(0.1)
        
        car_Info = driver.find_element_by_xpath(f'//*[@id="sr_normal"]/tr[{j}]/td[2]/a/span[2]/strong')
        car_info.append(car_Info.text)
        time.sleep(0.1)
        
        car_Year = driver.find_element_by_xpath(f'//*[@id="sr_normal"]/tr[{j}]/td[2]/span[1]/span[1]')
        car_year.append(car_Year.text)
        time.sleep(0.1)
   
        car_Km = driver.find_element_by_xpath(f'//*[@id="sr_normal"]/tr[{j}]/td[2]/span[1]/span[2]')
        car_km.append(car_Km.text)
        time.sleep(0.1)

        
        car_Location =  driver.find_element_by_xpath(f'//*[@id="sr_normal"]/tr[{j}]/td[2]/span[1]/span[4]')
        car_location.append(car_Location.text)
        time.sleep(0.1)
        
        car_Price = driver.find_element_by_xpath(f'//*[@id="sr_normal"]/tr[{j}]/td[3]/strong')
        car_price.append(car_Price.text)
        time.sleep(0.1)
        
        G_DD =  driver.find_element_by_xpath(f'//*[@id="sr_normal"]/tr[{j}]/td[2]/span[1]/span[3]')
        G_D.append(G_DD.text)
        
```

```python
WHITEdf2 = pd.DataFrame(
    {'brand': car_brand,
     'name':car_name,
     'year': car_year,
     'info': car_info,
     'km':car_km,
     'location':car_location,
     'price':car_price,
     'type':G_D
    })
```

```python
WHITEdf2['color'] = '흰색'
WHITEdf2['websites'] = 'Ncar'
```

```python
WHITEdf2.to_csv('white_N1.csv')
```
